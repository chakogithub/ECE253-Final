```python
import os
import cv2
import torch
import numpy as np
from tqdm import tqdm
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim

import sys
sys.path.append("../PyTorch-Image-Dehazing")

from net import dehaze_net
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

model = dehaze_net().to(device)
ckpt_path = "../PyTorch-Image-Dehazing/snapshots/dehazer.pth"

model.load_state_dict(torch.load(ckpt_path, map_location=device))
model.eval()

print("AOD-Net loaded.")
```

```python
def aod_dehaze(img):
    """img: uint8 RGB image, shape (H, W, 3)"""
    x = img.astype(np.float32) / 255.0
    x = torch.from_numpy(x).permute(2,0,1).unsqueeze(0).to(device)

    with torch.no_grad():
        out = model(x)

    out = out.squeeze().cpu().permute(1,2,0).numpy()
    out = np.clip(out*255, 0, 255).astype(np.uint8)
    return out
```

```python
SOTS_OUT_HAZY = "../data/RESIDE/SOTS/outdoor/hazy"
SOTS_OUT_GT   = "../data/RESIDE/SOTS/outdoor/gt"
RESULT_AOD_OUT = "../results/aod/outdoor"

os.makedirs(RESULT_AOD_OUT, exist_ok=True)
```

```python
results = []

hazy_files = sorted(os.listdir(SOTS_OUT_HAZY))
print("Total outdoor images:", len(hazy_files))

for fname in tqdm(hazy_files):
    if not fname.lower().endswith(("jpg","png")):
        continue

    # hazy path
    hpath = os.path.join(SOTS_OUT_HAZY, fname)

    # gt file based on first 4 digits
    base = fname[:4]
    gt_name = base + ".png"
    gt_path = os.path.join(SOTS_OUT_GT, gt_name)

    if not os.path.exists(gt_path):
        print("GT missing for:", fname)
        continue

    # read images
    hazy = cv2.imread(hpath)
    hazy = cv2.cvtColor(hazy, cv2.COLOR_BGR2RGB)

    gt = cv2.imread(gt_path)
    gt = cv2.cvtColor(gt, cv2.COLOR_BGR2RGB)

    # run AOD
    J = aod_dehaze(hazy)

    # save result
    out_path = os.path.join(RESULT_AOD_OUT, fname)
    cv2.imwrite(out_path, cv2.cvtColor(J, cv2.COLOR_RGB2BGR))

    # compute metrics
    p = psnr(gt, J)
    s = ssim(gt, J, channel_axis=2)

    results.append([fname, gt_name, p, s])
```

```python
import pandas as pd

df = pd.DataFrame(results, columns=["hazy_file","gt_file","psnr","ssim"])
print(df.head())

print(df[["psnr","ssim"]].describe())
```

```python
import matplotlib.pyplot as plt

# 根路径（根据你之前的结构，如果不同就改一下）
SOTS_OUT_HAZY = "../data/RESIDE/SOTS/outdoor/hazy"
DCP_OUT_DIR   = "../results/dcp/outdoor"
AOD_OUT_DIR   = "../results/aod/outdoor"

def show_compare(filename):
    """filename 例如 '0001_0.8_0.2.jpg'"""
    hazy_path = os.path.join(SOTS_OUT_HAZY, filename)
    dcp_path  = os.path.join(DCP_OUT_DIR, filename)
    aod_path  = os.path.join(AOD_OUT_DIR, filename)

    # 读取并转 RGB
    hazy = cv2.imread(hazy_path)
    hazy = cv2.cvtColor(hazy, cv2.COLOR_BGR2RGB)

    dcp = cv2.imread(dcp_path)
    dcp = cv2.cvtColor(dcp, cv2.COLOR_BGR2RGB)

    aod = cv2.imread(aod_path)
    aod = cv2.cvtColor(aod, cv2.COLOR_BGR2RGB)

    # 画图
    plt.figure(figsize=(12,4))
    plt.subplot(1,3,1)
    plt.imshow(hazy)
    plt.title("Hazy")
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.imshow(dcp)
    plt.title("DCP")
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.imshow(aod)
    plt.title("AOD-Net")
    plt.axis("off")

    plt.tight_layout()
    plt.savefig(filename)
    plt.show()

# 示例：看几张典型图
show_compare("0001_0.8_0.2.jpg")
show_compare("0002_0.8_0.08.jpg")
show_compare("0003_0.8_0.2.jpg")
```

```python
# Indoor

# SOTS INDOOR 数据路径（和之前 DCP 一样）
SOTS_IN_HAZY = "../data/RESIDE/SOTS/indoor/hazy"
SOTS_IN_GT   = "../data/RESIDE/SOTS/indoor/gt"

# AOD 结果输出目录
RESULT_AOD_IN = "../results/aod/indoor"
os.makedirs(RESULT_AOD_IN, exist_ok=True)
```

```python
hazy_files_in = sorted(
    [f for f in os.listdir(SOTS_IN_HAZY) 
     if f.lower().endswith(".png") and f.endswith("_10.png")]
)

print("Total indoor hazy images (level 10):", len(hazy_files_in))
print(hazy_files_in[:5])
```

```python
results_in = []

for fname in tqdm(hazy_files_in):
    hpath = os.path.join(SOTS_IN_HAZY, fname)

    # GT 文件名：去掉 "_10" 后缀
    base = fname.split("_")[0]   # '1400_10.png' → '1400'
    gt_name = base + ".png"
    gt_path = os.path.join(SOTS_IN_GT, gt_name)

    if not os.path.exists(gt_path):
        print("GT missing for:", fname)
        continue

    # 读图 (RGB)
    hazy = cv2.imread(hpath)
    hazy = cv2.cvtColor(hazy, cv2.COLOR_BGR2RGB)

    gt = cv2.imread(gt_path)
    gt = cv2.cvtColor(gt, cv2.COLOR_BGR2RGB)

    # AOD 去雾
    J = aod_dehaze(hazy)

    # 存结果
    out_path = os.path.join(RESULT_AOD_IN, fname)
    cv2.imwrite(out_path, cv2.cvtColor(J, cv2.COLOR_RGB2BGR))

    # 指标
    p = psnr(gt, J)
    s = ssim(gt, J, channel_axis=2)

    results_in.append([fname, gt_name, p, s])
```

```python
df_in = pd.DataFrame(results_in, columns=["hazy_file","gt_file","psnr","ssim"])
print(df_in.head())

summary_in = df_in[["psnr","ssim"]].describe()
print(summary_in)
```

```python
import cv2
import numpy as np
import torch

def dehaze_infer(image_path, net, device='cpu'):
    """
    输入文件路径，输出 RGB uint8 numpy array 的去雾图像
    """
    # 读取 BGR → 转 RGB → [0,1]
    img_bgr = cv2.imread(image_path)
    if img_bgr is None:
        raise FileNotFoundError(image_path)

    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB).astype(np.float32) / 255.

    # HWC → CHW → Tensor
    inp = torch.from_numpy(img_rgb).permute(2, 0, 1).unsqueeze(0).to(device)

    with torch.no_grad():
        out = net(inp)

    # 输出可能是 [0,1] 或 [-1,1]，依实际模型而定
    out = out.squeeze(0).permute(1, 2, 0).cpu().numpy()

    # 若网络输出范围是 [-1,1]，则修正：
    if out.min() < 0:
        out = (out + 1.0) / 2.0  

    out = np.clip(out * 255, 0, 255).astype(np.uint8)
    return out
```

```python
test_img = "../data/OHAZE/hazy/01_outdoor_hazy.jpg"  # 改成你的路径
out = dehaze_infer(test_img, net, device)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(cv2.imread(test_img), cv2.COLOR_BGR2RGB))
plt.title("Hazy")

plt.subplot(1,2,2)
plt.imshow(out)
plt.title("Dehaze_net Output")
plt.show()
```
